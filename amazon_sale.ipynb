```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/amazon.csv")
```

```python
#What is the average rating for each product category?
a=pd.to_numeric(df['rating'],errors='coerce').mean()
print(f'avarage rating is {a}')
```

```python
#q What are the top rating_count products by category?
df['rating_count']=pd.to_numeric(df['rating_count'],errors='coerce')
df[df['rating_count']==df['rating_count'].max()]
```

```python
#' What is the distribution of discounted prices vs. actual prices?
#df['actual_price']=df['actual_price'].str.replace(',','').str.replace('₹','')
#df['discounted_price']=df['discounted_price'].str.replace(',','').str.replace('₹','')
df['discounted_price']=df['discounted_price'].astype(float)
df['actual_price']=df['actual_price'].astype(float)
df['diff']=df['actual_price']-df['discounted_price']
df[['product_name','discounted_price','actual_price','diff']]
```

```python
#+' How does the average discount percentage vary across categoriesk
df['discount_percentage']=df['discount_percentage'].str.replace('%',"")
df['discount_percentage']=df['discount_percentage'].astype(float)

df.groupby(['category'])['discount_percentage'].mean()
```

```python
#' What are the most popular product namesk
#for most popular we can see the no. of highest rating
df['rating']=pd.to_numeric(df['rating'],errors='coerce')
aa=df.sort_values(by='rating', ascending=False).head()
aa[['product_id','rating','actual_price']]
```

```python
#What are the most popular product keywordsk
df
keywords = df['category'].str.lower().str.split('|')# convert all string lower case to avoid error and split it by |
all_keywords = keywords.explode() # explode will seperate str value
all_keywords.value_counts()
```
