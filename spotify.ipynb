```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/spotify.csv')
df
```

```python
#Read the dataframe, check null value if present then do the needful, check duplicate row , if present then do the needful
df[df.duplicated()==True]
```

```python
df=df.drop_duplicates()
df
```

```python
df.info()
```

```python
# What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram
cc=df['Popularity'].value_counts().reset_index()
cc
```

```python
px.histogram(cc,x='Popularity',y='count',nbins=100)
```

```python
#7 Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot
corrr=df[['Popularity','Duration (ms)']]
corrr.corr()
# positively corelated
```

```python
sns.scatterplot(data=df,x='Popularity',y='Duration (ms)')
sns.kdeplot(data=df,x='Popularity',y='Duration (ms)')
```

```python
#Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot
ss=df['Artist'].value_counts().reset_index()
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Artist', palette='Set2')
plt.title('Artists by Track Count')
plt.xlabel('Artist')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
df['Artist'].value_counts().head(1)
```

```python
#What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each
df.sort_values(by='Popularity',ascending=True).head(5)
```

```python
# Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist
plt.figure(figsize=(20,8))
sns.barplot(data=df,x='Artist',y='Popularity')
plt.xticks(rotation=90)
plt.show()
```

```python
aa=df.groupby(['Artist'])['Popularity'].mean()
df.drop('popularity_mean',axis=1,inplace=True)
df['Artist_popularity_mean']=df['Artist'].map(aa)
bb=df.sort_values(by='Popularity',ascending=False).head(10)
display(bb)
print('the artist among top 10 have most popularity is ')
bb[bb['Artist_popularity_mean']==bb['Artist_popularity_mean'].max()]
```

```python
# Visualize relationships between multiple numerical variables simultaneously using a pair plot
sns.pairplot(df)
```

```python
#Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot
plt.figure(figsize=(20,6))
sns.violinplot(data=df,x='Artist',y='Duration (ms)')
plt.xticks(rotation=90)
plt.show()
```

```python
#How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.
```

```python
plt.figure(figsize=(20, 6))
sns.swarmplot(x='Artist', y='Popularity', data=df,hue='Duration (ms)')
plt.title('Track Popularity Distribution (Swarm Plot)')
plt.xticks(rotation=90)
plt.show()
```
